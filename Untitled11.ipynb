```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
import math

# Download data
ticker = "AAPL"
df = yf.download(ticker, start="2010-01-01", end="2024-01-01")
if df.empty:
    raise ValueError("Error: No data downloaded for the given ticker and timeframe.")

# Use closing price and drop missing values
df = df[['Close']].dropna()

# Scale data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Create sequences
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(sequence_length, len(data)):
        X.append(data[i-sequence_length:i])
        y.append(data[i])
    return np.array(X), np.array(y)

sequence_length = 60
X, y = create_sequences(scaled_data, sequence_length)

# Split into training and test sets (80% train, 20% test)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Reshape input to be [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=3)

# Train the model
model.fit(X_train, y_train, validation_split=0.1, epochs=20, batch_size=32, callbacks=[early_stop], verbose=1)

# Predict
predictions = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predictions)
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluate performance
rmse = math.sqrt(mean_squared_error(actual_prices, predicted_prices))
print(f"Test RMSE: {rmse:.2f}")

# Plot
plt.figure(figsize=(14, 7))
plt.plot(df.index[-len(actual_prices):], actual_prices, label='Actual Price')
plt.plot(df.index[-len(predicted_prices):], predicted_prices, label='Predicted Price')
plt.title('AAPL Stock Price Prediction with LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
!pip install gradio
import gradio as gr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
import math

# Define the prediction function
def predict_lstm(ticker, start_date, end_date):
    """Predicts stock prices using LSTM and returns evaluation and plot."""
    try:
        # Download data
        df = yf.download(ticker, start=start_date, end=end_date)
        if df.empty:
            raise ValueError("Error: No data downloaded for the given ticker and timeframe.")

        # Use closing price and drop missing values
        df = df[['Close']].dropna()

        # Scale data
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(df)

        # Create sequences
        def create_sequences(data, sequence_length):
            X, y = [], []
            for i in range(sequence_length, len(data)):
                X.append(data[i - sequence_length:i])
                y.append(data[i])
            return np.array(X), np.array(y)

        sequence_length = 60
        X, y = create_sequences(scaled_data, sequence_length)

        # Split into training and test sets (80% train, 20% test)
        split = int(0.8 * len(X))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        # Reshape input to be [samples, time steps, features]
        X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
        X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

        # Build LSTM model
        model = Sequential()
        model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
        model.add(LSTM(units=50))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Early stopping
        early_stop = EarlyStopping(monitor='val_loss', patience=3)

        # Train the model
        model.fit(X_train, y_train, validation_split=0.1, epochs=20, batch_size=32, callbacks=[early_stop], verbose=0) # Set verbose to 0 to suppress training output

        # Predict
        predictions = model.predict(X_test)
        predicted_prices = scaler.inverse_transform(predictions)
        actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

        # Evaluate performance
        rmse = math.sqrt(mean_squared_error(actual_prices, predicted_prices))
        evaluation_text = f"Test RMSE: {rmse:.2f}"

        # Plot
        plt.figure(figsize=(14, 7))
        plt.plot(df.index[-len(actual_prices):], actual_prices, label='Actual Price')
        plt.plot(df.index[-len(predicted_prices):], predicted_prices, label='Predicted Price')
        plt.title(f'{ticker} Stock Price Prediction with LSTM')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        # Save plot to a temporary file
        plt.savefig('prediction_plot.png')
        plt.close() # Close the plot to release resources

        return evaluation_text, 'prediction_plot.png' # Return the path to the saved plot

    except ValueError as e:
        return str(e), None  # Return error message and None for the image if data download fails

# Gradio interface
interface = gr.Interface(
    fn=predict_lstm,
    inputs=[
        gr.Textbox(label="Stock Ticker", value="AAPL"),
        gr.Textbox(label="Start Date (YYYY-MM-DD)", value="2010-01-01"),
        gr.Textbox(label="End Date (YYYY-MM-DD)", value="2024-01-01")
    ],
    outputs=[
        gr.Textbox(label="Model Evaluation"),
        gr.Image(label="Prediction Plot")
    ],
    title="LSTM Stock Price Predictor",
    description="Predicts and compares stock prices using LSTM based on historical Yahoo Finance data."
)

interface.launch()
```
